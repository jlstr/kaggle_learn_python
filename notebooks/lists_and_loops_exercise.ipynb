```python
def has_lucky_number(nums):
  """Return whether the given list of numbers is lucky. A lucky list contains
  at least one number divisible by 7.
  """
  # return len([num % 7 == 0 for num in nums]) > 0
  return len([num for num in nums if num % 7 == 0]) > 0
```

```python
nums = [1, 2, 4, 7]
has_lucky_number(nums)
```

```python
nums = [3]
has_lucky_number(nums)
```

```python
def elementwise_greater_than(L, thresh):
  """Return a list with the same length as L, where the value at index i is 
  True if L[i] is greater than thresh, and False otherwise.
  
  >>> elementwise_greater_than([1, 2, 3, 4], 2)
  [False, False, True, True]
  """

  return [num > thresh for num in L]
```

```python
elementwise_greater_than([1, 2, 3, 4], 2)
```

```python
def menu_is_boring(meals):
  """Given a list of meals served over some period of time, return True if the
  same meal has ever been served two days in a row, and False otherwise.
  """
  prev_meal = None

  for meal in meals:
    if prev_meal and prev_meal == meal:
      return True

    prev_meal = meal

  return False

meals = ["hornado", "perro", "shawarma", "pizza", "hamburguesa"]

print(menu_is_boring(meals))
```
